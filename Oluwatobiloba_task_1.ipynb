Exercise 1: Create a Student Class


Create a Student class with the following requirements:
- Attributes: name, student_id, grades (list)
- Methods: add_grade(), get_average(), get_status()
- get_status() should return "Pass" if average >= 60, otherwise "Fail"

```python
# Exercise 1

class student:
    
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = []
        
    def add_grades(self, grade):
        self.grades.append(grade)
        
    def calculate_average(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)

    def get_status(self):
        if self.calculate_average() >= 60:
            return "pass"
        else:
            return "fail"    
        
        
        
```

```python
# create student

student1 = student("Salma", "evol")

student1.add_grades(90)
student1.add_grades(87)
student1.add_grades(88)
    

print(student1.name)
print(student1.student_id)
print(student1.grades)
print(student1.calculate_average())
print(student1.get_status())
```

### Exercise 2: Error Handling for User Input
Create a function that safely gets user input and converts it to an integer:
- Keep asking until valid input is provided
- Handle ValueError for non-numeric input
- Handle KeyboardInterrupt when user presses Ctrl+C
- Return the valid integer

```python
def get_integer():
    while True:
        try:
            number = int(input("enter an integer: "))
            return number
        
        except ValueError:
            print("invalid input! please enter numbers only.")
            
        except KeyboardInterrupt:
            print("\nProgram interrupted by user.")
            return None      
        
        value = get_integer()
        print("You entered:", value)
        
```

### Exercise 3: Library Management System
Create a simple library system with error handling:
- Book class with title, author, is_available attributes
- Library class that manages multiple books
- Methods: add_book(), borrow_book(), return_book()
- Include appropriate error handling for:
  - Trying to borrow a book that doesn't exist
  - Trying to borrow a book that's not available
  - Trying to return a book that wasn't borrowed

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True
        
class Library:
    def __init__(self):
        self.books = []
        
    def add_book(self, book):
        self.books.append(book)
        print(f"{book.title} has been added to the library.")
        
    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                
                if not book.is_available:
                    raise Exception("This book is currently not available.")
                
                book.is_available = False
                print(f" you borrowed '{book.title}'.")
                return
            
        raise Exception("book does not exist in the library.")
    
    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                
                if book.is_available:
                    raise Exception("This book was not borrowed.")
                
                book.is_available = True
                print(f"you returned '{book.title}'.")
                return
            
        raise Exception("book does not exist in the library.")
```

```python
library = Library()

book1 = Book("Sedimentology", "Chris Brown")
book2 = Book("Sedimentary Basins", "David Adeleke")

library.add_book(book1)
library.add_book(book2)

try:
    library.borrow_book("Sedimentology")
except Exception as error:
    print(error)
    
try:
    library.return_book("Sedimentary Basins")
except Exception as error:
    print(error)
```
