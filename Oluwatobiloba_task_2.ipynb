```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)


# Create a rectangle
my_rectangle = Rectangle(5, 10)

print(f"Area: {my_rectangle.area()}")
print(f"Perimeter: {my_rectangle.perimeter()}")
```

```python
def get_integer_input(user_input):
    try:
        # Convert string to integer
        result = int(user_input)
        return f"Success! Your integer is: {result}"
    except ValueError:
        # Handle the error if conversion fails
        return f"Error: '{user_input}' is not a valid number."


# Example 1: Valid input
print(get_integer_input("42"))

# Example 2: Invalid input
print(get_integer_input("hello"))
```
